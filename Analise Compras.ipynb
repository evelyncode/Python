```python
import pandas as pd
import numpy as np
```

```python
df="dados_compras.json"
```

```python
dados= pd.read_json(df,orient= "records")
dados.head()
```

```python
#informações sobre os compradores: usar a função loc para criar uma tabela apenas com
#informações dos compradores e inciar a avaliação
```

```python
dados_compradores=dados.loc[:,['Idade','Sexo','Login']]
dados_compradores.head()
```

```python
#Como o mesmo comprador pode aparecer várias vezes, precisamos remover os duplicados
dados_compradores=dados_compradores.drop_duplicates()
contagen_compradores=dados_compradores.count()[0]
contagen_compradores
#usa o 0 no count para mostrar apenas o totais de dados unicos e não por coluna
```

```python
#converter em dataframe
pd.DataFrame({"Total de jogadores":[contagen_compradores]}) #chaveValor
```

```python
#Analise da compra - Calculos Basicos
media_compras=dados['Valor'].mean()
valor_compras=dados['Valor'].sum()
total_compras=dados['Valor'].count()
itens_unicos=len(dados['Item ID'].unique())
```

```python
#trasformar em um dataframe

calculos_basicos = pd.DataFrame({'Media dos dados' :[media_compras],
                                 'Soma das compras' : valor_compras,
                                 'Valor Total das compras' :total_compras,
                                 'Itens Unicos':itens_unicos})
calculos_basicos
```

```python
#Data Munging = Formatação dos dados
```

```python
calculos_basicos.round(2)
calculos_basicos
```

```python
calculos_basicos['Media dos dados']=calculos_basicos['Media dos dados'].map('${:,.2f}'.format)
calculos_basicos['Soma das compras']=calculos_basicos['Soma das compras'].map('${:,.2f}'.format)
calculos_basicos
```

```python
#itens unicos
dados['Item ID'].unique()
```

```python
#Calcular a quantidade de pessoas por sexo e o percentual
```

```python
#função value_conts, conta os valores diferentes de uma coluna de um ddf
contagem_sexo= dados_compradores['Sexo'].value_counts()
contagem_sexo
```

```python
perc_sexo= (contagem_sexo/contagen_compradores)*100
dados_genero= pd.DataFrame({'Sexo': contagem_sexo, '%': perc_sexo})
dados_genero
```

```python
#data munging --- nada mais é que formatar os dados. A função rount 2 arredonda as casas decimais
dados_genero = dados_genero.round(2)
dados_genero
```

```python
dados_genero['%'] =dados_genero['%'].map('{:,.1f}%'.format)
dados_genero
```

```python
#analise de compras por genero - 1: Numero de compras por sexo,
#Valor médio de item, total, total normalidado
#usei o groupby para agrupar por sexo, e realizei a soma de valor
total_genero= dados.groupby(['Sexo']).sum()['Valor'].rename('Total de vendas')
media_genero= dados.groupby(['Sexo']).mean()['Valor'].rename('Media de preço')
compras_genero= dados.groupby(['Sexo']).count()['Valor'].rename('Total de compras')
media_ajustado=total_genero/dados_genero['Sexo']
media_ajustado
media_ajustado =media_ajustado.round(2)
```

```python
media_genero=media_genero.round(2)
media_genero
```

```python
df_genero= pd.DataFrame({'Total de compras':compras_genero,
                        'Media de preço':media_genero,
                        'Media Ajustada':media_ajustado,
                        'Valor Total': total_genero})
#data Munging

df_genero['Media de preço'] =df_genero['Media de preço'].map('${:,.2f}'.format)
df_genero['Media Ajustada'] =df_genero['Media Ajustada'].map('${:,.2f}'.format)
df_genero['Valor Total']=df_genero['Valor Total'].map('${:,.2f}'.format)
df_genero
```

```python
#Continuando
dados
```

```python
coluna_idade = [0,9.99,19.99,29.99,39.99,49.99,999]
nomes_col_idade=['Menos de 10','10 a 19','20 a 29','30 a 39','40 a 49','Acima de 50 anos']
```

```python
#pandas cut reliza corte de uma lista com base nos parametros
dados['Range Idades']=pd.cut(dados_compradores['Idade'], coluna_idade,
                                         labels=nomes_col_idade)
dados['Range Idades']
```

```python
#calculos básicos para alimentar o df
cont_dados=dados['Range Idades'].value_counts()
mean_dados=dados.groupby(['Range Idades']).mean()['Valor']
sum_dados=dados.groupby(['Range Idades']).sum()['Valor']
per_dados=(cont_dados/contagen_compradores)*100
per_valor=(sum_dados/valor_compras)*100

#dataframe

calc_dados=pd.DataFrame({'Contagem': cont_dados, 'Média': mean_dados,'Valor Total':sum_dados,
                        '% Compradores':per_dados, '% Valor': per_valor})
calc_dados=calc_dados.round(2)
calc_dados

#Data Munging
calc_dados['Média']=calc_dados['Média'].map("${:,.2f}".format)
calc_dados['Valor Total']=calc_dados['Valor Total'].map('${:,.2f}'.format)
calc_dados['% Compradores']=calc_dados['% Compradores'].map('%{:,.2f}'.format)
calc_dados['% Valor']=calc_dados['% Valor'].map('%{:,.2f}'.format)
calc_dados
```

```python
user_total=dados.groupby(['Login']).sum()['Valor'].rename("Valor Total")
user_mean=dados.groupby(['Login']).mean()['Valor'].rename('Média')
user_count=dados.groupby(['Login']).count()['Valor'].rename('Total Itens')
user_per=(user_count/contagen_compradores)*100
```

```python
df_user=pd.DataFrame({'Valor Total': user_total,'Média':user_mean,
                     'Total itens':user_total,'%':user_per})
df_user=df_user.round(2)
df_user['Valor Total']=df_user['Valor Total'].map('${:,.2f}'.format)
df_user['Média']=df_user['Média'].map('${:,.2f}'.format)
df_user['%']=df_user['%'].map('%{:,.2f}'.format)
df_user
```

```python
#fazer o mesmo para ítens
itens_sum=dados.groupby(['Nome do Item']).sum()['Valor']
itens_mean=dados.groupby(['Nome do Item']).mean()['Valor']
itens_total=dados.groupby(['Nome do Item']).count()['Valor']
itens_perc=(itens_total/itens_unicos)*100

df_itens=pd.DataFrame({'Valor Total':itens_sum,'Média':itens_mean,
                      'Total itens':itens_total,'%':itens_perc})
df_itens.round(2)
df_itens['%']=df_itens['%'].map('%{:,.2f}'.format)
df_itens['Valor Total']=df_itens['Valor Total'].map('${:,.2f}'.format)
df_itens['Média']=df_itens['Média'].map('${:,.2f}'.format)
df_itens

```
